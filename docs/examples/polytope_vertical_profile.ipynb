## Polytope vertical profile

The following example first retrieves vertical profiles on pressure levels using polytope, then converts the resulting CoverageJSON data to Xarray.

```python
import earthkit.data as ekd

request = {
    "class": "od",
    "stream": "enfo",
    "type": "pf",
    "date": "20241114",
    "time": "0000",
    "levtype": "pl",
    "expver": 1,
    "domain": "g",
    "param": "203/133",
    "number": "1",
    "step": "0",
    "levelist": "1/to/1000",
    "feature": {
        "type": "verticalprofile",
        "points": [[38.9, -9.1]],
    },
}

ds = ekd.from_source("polytope", "ecmwf-mars-beta", request, stream=False, address='polytope-test.ecmwf.int')
ds.to_xarray()
```
