### Quickplot

With **earthkit-plots**, it's super easy to visualise data that you've loaded with **earthkit-data**. Let's start by getting a local file:

```python
data = ek.data.from_source("file", "era5-monthly-mean-2t-199312.nc")
```

The easiest way to visualise this is with the `quickplot()` function. This produces a quick visualisation of your data with some nice automatic features, like smart titles and labels populated from your data's metadata, plus high-quality default styles applied based on the variable plotted. For example in this case we have temperature data, for which earthkit-plots has built-in default styles:

```python
ek.plots.quickplot(data)
```

You can also quickly and easily convert the units of your data on-the-fly, by simply passing the name of the units to quickplot - for example with temperature we can plot in Kelvin, celsius or Fahrenheit. You can also pass a named or custom domain, and **earthkit-plots** will do the rest, calculating the optimal projection for your domain. It's that simple.

```python
ek.plots.quickplot(data, domain="Europe", units="celsius")
```

### Going deeper

If you want more control, you can go much deeper into layouts, components and styles.

```python
chart = ek.plots.Map(domain=["France", "Spain"])

chart.contourf(data, colors="Spectral_r", levels=range(-12, 21), units="celsius")
chart.legend()
chart.coastlines()
chart.borders()

chart.title("ERA5 monthly average {variable_name} over {domain} - {time:%B %Y}")

chart.show()
```

### Complex grids

Another powerful feature is support for complex grids - like HEALPix or Octahedral grids - thanks to **earthkit-regrid** working seamlessly in the background.

```python
data = ek.data.from_source("sample", "healpix-h128-nested-2t.grib")
data.ls()
```

You can pass this kind of data directly to **earthkit-plots**, and it‚Äôll handle everything, with no manual regridding required.

```python
ek.plots.quickplot(data, units="celsius", domain="Africa")
```
