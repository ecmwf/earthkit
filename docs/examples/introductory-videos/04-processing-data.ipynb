### Simple operations

When you load data using **earthkit-data**, you can perform standard arithmetic operations - like addition, subtraction, multiplication, and division - directly on your data.

Let's start by getting some data:

```python
import earthkit as ek

data = ek.data.from_source("file", "era5-monthly-mean-2t-199312.grib")
```

But the earthkit ecosystem also includes specialised tools for common analysis and processing tasks, exposed through a high-level, user-friendly API. For example, there's earthkit-transforms, which provides powerful methods for aggregating data across dimensions like time, space or dimensions of your choice.

There are also highly convenient pre-made functions for doing things like resampling your data to specific temporal resolutions like daily, monthly, seasonal or annual:

```python
daily_mean = ek.transforms.aggregate.temporal.daily_mean(data)
```

Plus methods for masking or aggerating gridded data over geospatial shapes:

```python
italy = ek.data.from_source("file", "italy.geojson")
italy_mask = ek.transforms.aggregate.spatial.mask(data, italy)

ek.plots.quickplot(italy_mask, domain="Italy")
```
